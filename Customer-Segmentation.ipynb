```python
A_I15  = df["Annual Income (k$)"] [(df["Annual Income (k$)"] >= 0) & (df["Annual Income (k$)"] <= 15)]
A_I15_40  = df["Annual Income (k$)"] [(df["Annual Income (k$)"] >= 15) & (df["Annual Income (k$)"] <= 40)]
A_I41_80  = df["Annual Income (k$)"] [(df["Annual Income (k$)"] >= 41) & (df["Annual Income (k$)"] <= 80)]
A_I81_140  = df["Annual Income (k$)"] [(df["Annual Income (k$)"] >= 81) & (df["Annual Income (k$)"] <= 140)]
A_I141_200  = df["Annual Income (k$)"] [(df["Annual Income (k$)"] >= 141) ]


yAnual = [len(A_I15.values),len(A_I15_40.values),len(A_I41_80.values), len(A_I81_140.values), len(A_I141_200.values)]
xAnual = ["$0-15k","$15k-40k","$40k-80k","$81k-140k","$141k+"]

plt.figure(figsize=(15,6))
sns.barplot (x=xAnual, y =yAnual, palette="mako") 
plt.title("Annual Income") 
plt.xlabel ("Income")
plt.ylabel ("Number of Customer Having the Income")
plt.show()
```

```python
x1 =df.loc[:, ["Age", "Spending Score (1-100)"]].values
from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit (x1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize= (12,6))
plt.grid
plt.plot(range (1,11), wcss, linewidth=2, color="red", marker ="8") 
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
label = kmeans.fit_predict (x1)
print(label) 
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter (x1[:,0], x1[:,1], c=kmeans.labels_, cmap = 'rainbow')
plt.scatter (kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], color='black') 
plt.title('Clusters of Customers') 
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
